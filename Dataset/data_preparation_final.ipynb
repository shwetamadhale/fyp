## Generating dataset


```python
header = ["label"]
import cv2
import csv
import glob
 
header  =["label"]
for i in range(0,784):
    header.append("pixel"+str(i))
with open('dataset_a.csv', 'a') as f:
    writer = csv.writer(f)
    writer.writerow(header)
   
               
```

## Reading images and assigning labels

```python
for label in range(135):
    dirList1 = glob.glob("C:/Users/madha/Downloads/trial2/data/"+str(label)+"/*.png")
    dirList2 = glob.glob("C:/Users/madha/Downloads/trial2/data/"+str(label)+"/*.jpg")
    dirList3 = glob.glob("C:/Users/madha/Downloads/trial2/data/"+str(label)+"/*.bmp")
    dirList = dirList1 + dirList2 + dirList3
    for img_path in dirList:
        im= cv2.imread(img_path)
        im_gray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
        im_gray = cv2.GaussianBlur(im_gray,(15,15), 0)
        roi= cv2.resize(im_gray,(28,28), interpolation=cv2.INTER_AREA)
        data=[]
        data.append(label)
        rows, cols = roi.shape
        ## Add pixel one by one into data array
        for i in range(rows):
            for j in range(cols):
                k =roi[i,j]
                data.append(k)
        with open('dataset.csv', 'a') as f:
            writer = csv.writer(f)
            writer.writerow(data)
 

     
```
